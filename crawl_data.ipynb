```python
import requests
import requests_cache
from bs4 import BeautifulSoup
import json
import time
import pandas as pd
import datetime as dt
import re
from selenium import webdriver
import urllib.robotparser
```

```python
sleep_time = 1
```

```python
headers = { 'X-Requested-With': 'XMLHttpRequest', "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36"}
```

```python
url = "https://tiki.vn/dien-thoai-may-tinh-bang/c1789?page={}&src=c.1789.hamburger_menu_fly_out_banner"
url1 = 'https://tiki.vn/api/v2/products?limit=48&include=advertisement&aggregations=1&category=8095&page={}&urlKey=laptop'
product_url = "https://tiki.vn/api/v2/products/{}"

product_list = []
product_detail_list = []
name_list = []
quantity_sold_list = []
i = 1
while (True):
    print("Crawl page: ", i)
    requests_cache.install_cache()
    print('url:', url1.format(i))
    r = requests.get(url1.format(i), headers=headers)
#     r = urllib.request.urlopen(url.format(i), headers=headers)
    if not r.from_cache: 
        time.sleep(sleep_time)
        r = requests.get(url1.format(i), headers=headers)
#         r = urllib.request.urlopen(url.format(i), headers=headers)
    if (r.status_code != 200):
        break
    tree = BeautifulSoup(r.text, features='html.parser')  
    product_box = tree.find_all('a', class_="product-item")
    if (len(product_box) == 0):
        break
    
    for product in product_box:
        # name of product
        name_tag = product.find(class_ = 'name')
        print(name_tag)
        name_list.append(name_tag.string)        
        
        # number of products sold
        quantity_sold = product.find('div', class_ = 'styles__StyledQtySold-sc-732h27-2')
        if quantity_sold == None:
            quantity_sold_list.append('0')
        else: quantity_sold_list.append(quantity_sold.string)
        break
    i += 1
```
